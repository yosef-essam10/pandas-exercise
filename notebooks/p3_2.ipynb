```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AgeDataset-V1.csv')
```

```python
df.head(10)
```

```python
df.columns
```

```python
df.groupby('Occupation')['Age of death'].mean()
```

```python
gender_counts = df.groupby('Occupation')['Gender'].value_counts(normalize=True).unstack()
male_ratio = gender_counts['Male']
male_ratio_sorted = male_ratio.sort_values(ascending=False)

print(male_ratio_sorted)
```

```python
df.groupby('Occupation')['Age of death'].agg(['min', 'max'])
```

```python
df.groupby(['Occupation','Gender'])['Age of death'].mean()
```

```python
gender_percent = (
    df.groupby('Occupation')['Gender'].value_counts(normalize=True) .unstack().fillna(0) * 100  
)

print(gender_percent.round(2))
```
