```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
n = 30

data = pd.DataFrame({
    'preTestScore': np.random.randint(0, 100, size=n),
    'postTestScore': np.random.randint(0, 100, size=n),
    'age': np.random.randint(18, 60, size=n)
})
```

```python
data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(
    data['preTestScore'], 
    data['postTestScore'], 
    s=data['age']*5,  
    alpha=0.6,
    c='teal',
    edgecolors='black'
)

plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Pre vs Post Test Scores (Size by Age)')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x=data['preTestScore'], y=data['postTestScore'])
```

```python
data['sex'] = np.random.choice(['Male', 'Female'], size=len(data))
```

```python
data.head(3)
```

```python
colors = {'Male': 'blue', 'Female': 'red'}

plt.figure(figsize=(10, 6))
plt.scatter(
    data['preTestScore'],
    data['postTestScore'],
    s=data['postTestScore'] * 4.5,  
    c=data['sex'].map(colors),     
    alpha=0.6,
    edgecolors='black'
)

plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Test Scores by Sex (Size by Post-Test Score)')
plt.grid(True)
plt.show()
```
