```python
import pandas as pd
import numpy as np
```

```python
crime = pd.read_csv('US_Crime_Rates_1960_2014.csv')
```

```python
crime.head(20)
```

```python
crime.info()
```

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
```

```python
crime.head(3)
```

```python
crime = crime.drop(columns=['Total'])
```

```python
crime.head(3)
```

```python
crime['Decade'] = (crime['Year'].dt.year // 10) * 10
numeric_cols = crime.select_dtypes(include='number').drop(columns='Population')
numeric_cols['Decade'] = crime['Decade']
df_grouped = numeric_cols.groupby('Decade').sum()
```

```python
df_grouped
```

```python
crime.head(3)
```

```python
df_grouped['Total_Crimes'] = df_grouped['Violent'] + df_grouped['Property']
most_dangerous_decade = df_grouped['Total_Crimes'].idxmax()
highest_value = df_grouped['Total_Crimes'].max()

print(f"The most dangerous decade to live in the US was the {most_dangerous_decade}s with {highest_value:,} reported crimes.")
```
