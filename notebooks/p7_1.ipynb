```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
chipo = pd.read_csv('chipotle.tsv', sep='\t')
```

```python
chipo.head(10)
```

```python
x = chipo.item_name
letter_counts = Counter(x)
df = pd.DataFrame.from_dict(letter_counts, orient='index')
df = df[0].sort_values(ascending = True)[45:50]

df.plot(kind='bar')

plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')

plt.show()
```

```python
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price] 

orders = chipo.groupby('order_id').sum()

plt.scatter(x = orders.item_price, y = orders.quantity, s = 50, c = 'green')

plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
```

##

create pie chart to item_name

```python
item_name_counts = chipo['item_name'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(item_name_counts, labels=item_name_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of chipotle item_name')
plt.axis('equal')
plt.show()
```
