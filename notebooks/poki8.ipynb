```python
import pandas as pd
import numpy as np
```

```python
data = {
    "name": ["Pikachu", "Charmander", "Bulbasaur"],
    "type": ["Electric", "Fire", "Grass/Poison"],
    "hp": [35, 39, 45],
    "evolution": ["Raichu", "Charmeleon", "Ivysaur"],
    "pokedex": [25, 4, 1]
}

columns_order = ["name", "type", "hp", "evolution", "pokedex"]
df = pd.DataFrame(data, columns=columns_order)

display(df)
```

```python
df["place"] = ["Viridian Forest", "Rocky Mountains", "Pallet Town Fields"]
```

```python
display(df)
```

####

merge data with other data frame

```python
data2 = {
    "name": ["Squirtle", "Jigglypuff", "Meowth"],
    "type": ["Water", "Normal/Fairy", "Normal"],
    "hp": [44, 115, 40],
    "evolution": ["Wartortle", "Wigglytuff", "Persian"],
    "pokedex": [7, 39, 52],
    "place": ["Lakes and Rivers", "Moonlight Fields", "Dark Alleys"]
}

df2 = pd.DataFrame(data2, columns=["name", "type", "hp", "evolution", "pokedex", "place"])
display(df2)
```

```python
df3 = pd.merge(df, df2, on="name",how='outer')
df3
```
