```python
import pandas as pd
import numpy as np
from datetime import datetime
import pandas_datareader.data as web
```

```python
start = pd.to_datetime("2015-01-01")
end = pd.to_datetime(datetime.today().date())
date_range = pd.date_range(start=start, end=end)
```

```python
api_key = 'your_alpha_vantage_key_here'
```

```python
df_apple = web.DataReader('AAPL', 'av-daily', start=start, end=end, api_key=api_key)
```

```python
df_apple
```

```python
df_apple["stock"] = "AAPL"
```

```python
df_apple.head()
```

```python
symbols = ["TSLA", "IBM", "MSFT"]

df_tsla = web.DataReader('TSLA', 'av-daily', start=start, end=end, api_key=api_key)
df_tsla["stock"] = "TSLA"

df_ibm = web.DataReader('IBM', 'av-daily', start=start, end=end, api_key=api_key)
df_ibm["stock"] = "IBM"

df_msft = web.DataReader('MSFT', 'av-daily', start=start, end=end, api_key=api_key)
df_msft["stock"] = "MSFT"
```

```python
df = pd.concat([df_apple, df_tsla, df_ibm, df_msft])
df
```

```python
df = df.set_index('stock', append=True)
df = df.swaplevel(0, 1)
```

```python
df
```

```python
vol = df[["volume"]]
```

```python
vol
```

```python
vol.index.set_names(['stock', 'date'], inplace=True)
```

```python
vol.head(5)
```

```python
vol_reset = vol.reset_index()


vol_reset['date'] = pd.to_datetime(vol_reset['date'])
weekly_vol = (
    vol_reset
    .groupby('stock')                       
    .resample('W', on='date')                
    .sum(numeric_only=True)                 
    .reset_index()                           
)
```

```python
weekly_vol
```

```python
vol_2015 = vol_reset[vol_reset['date'].dt.year == 2015]
total_volume_2015 = vol_2015['volume'].sum()
print("Total volume traded in 2015:", total_volume_2015)
```

```python
volume_per_stock_2015 = vol_2015.groupby('stock')['volume'].sum()
print(volume_per_stock_2015)
```
