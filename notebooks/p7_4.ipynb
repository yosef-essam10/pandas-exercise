```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=pd.read_csv('tips.csv')
```

```python
tips
```

```python
tips.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
tips.head(3)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(tips['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=tips, 
    x='total_bill', 
    y='tip', 
    size='size',    
    hue='size',     
    palette='viridis',
    sizes=(20, 200), 
    alpha=0.7
)

plt.title('Relationship between Total Bill, Tip, and Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Size', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['day'])
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x='tip',
    y='day',
    hue='sex',
    data=tips,
    palette={'Male': 'blue', 'Female': 'red'},
    s=100,   
    alpha=0.7
)

plt.title('Tip Amount by Day Differentiated by Sex')
plt.xlabel('Tip')
plt.ylabel('Day of the Week')
plt.legend(title='Sex')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(
    x='day',
    y='total_bill',
    hue='time',    
    data=tips,
    palette='Set2'
)

plt.title('Total Bill Distribution per Day by Time (Dinner vs Lunch)')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Time')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)


sns.histplot(tips[tips['time'] == 'Dinner']['tip'], bins=15, color='skyblue', ax=axes[0])
axes[0].set_title('Tip Distribution - Dinner')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frequency')


sns.histplot(tips[tips['time'] == 'Lunch']['tip'], bins=15, color='salmon', ax=axes[1])
axes[1].set_title('Tip Distribution - Lunch')
axes[1].set_xlabel('Tip')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True, sharex=True)


sns.scatterplot(
    x='total_bill', y='tip',
    hue='smoker',
    data=tips[tips['sex'] == 'Male'],
    palette={'Yes': 'red', 'No': 'green'},
    ax=axes[0],
    alpha=0.7,
    s=80
)
axes[0].set_title('Male: Total Bill vs Tip')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')
axes[0].legend(title='Smoker')


sns.scatterplot(
    x='total_bill', y='tip',
    hue='smoker',
    data=tips[tips['sex'] == 'Female'],
    palette={'Yes': 'red', 'No': 'green'},
    ax=axes[1],
    alpha=0.7,
    s=80
)
axes[1].set_title('Female: Total Bill vs Tip')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')
axes[1].legend(title='Smoker')

plt.tight_layout()
plt.show()
```
