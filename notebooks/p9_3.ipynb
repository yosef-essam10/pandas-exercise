```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv', parse_dates=['Date'], na_values=['', ' '])
```

```python
df
```

```python
date_diffs = df['Date'].diff().dropna()

# Most common time difference
most_common_freq = date_diffs.mode()[0]
print(f"Most common frequency: {most_common_freq}")
```

```python
inferred_freq = pd.infer_freq(df['Date'])
print(f"Inferred frequency: {inferred_freq}")
```

```python
df.set_index('Date', inplace=True)
```

```python
type(df.index)
```

```python
monthly = df.resample('ME').sum()
monthly
```

```python
monthly = monthly.dropna(how='all')
```

```python
monthly
```

```python
yearly = monthly.resample('YE').sum()
yearly
```

"In which year was the highest value of net financial flows (column 'Total') recorded, and what was that value?"

```python
df['year'] = df.index.year
df[df['year'] == 2016]['Total'].describe()
```

```python
yearly = yearly.dropna(subset=['Total'])  # نشيل السنين اللي فيها NaN فقط
max_year = yearly['Total'].idxmax()
max_value = yearly['Total'].max()
print(max_year, max_value)
```

```python
the_year=yearly['Total'].idxmax()   
value_the_year=yearly['Total'].max() 
print(f"highest value of net financial flows in year:{the_year} and value is {value_the_year}")
```

```python
valid_yearly = yearly[yearly['Total'] > 0]

if not valid_yearly.empty:
    max_year = valid_yearly['Total'].idxmax()
    max_value = valid_yearly['Total'].max()
    print(f"Year with highest net financial flows: {max_year.date()}, Value: {max_value}")
else:
    print("No valid positive Total values found in the dataset.")
```

```python
if not yearly['Total'].isna().all():
    max_year = yearly['Total'].idxmax()
    max_value = yearly['Total'].max()
    print(f"Year with least negative (or highest) net financial flows: {max_year.date()}, Value: {max_value}")
else:
    print("All yearly Total values are NaN.")
```
