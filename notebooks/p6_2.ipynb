```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('wind.data', delim_whitespace=True)
```

```python
df.head(3)
```

```python
def fix_year(yr):
    yr = int(yr)
    if yr < 100:         
        return 1900 + yr
    elif 1900 <= yr <= 2025: 
        return yr
    else:                
        return None


df['Yr'] = df['Yr'].apply(fix_year)


print(df['Yr'].unique())
```

```python
df
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df[['Yr', 'Mo', 'Dy']].rename(columns={
    'Yr': 'year',
    'Mo': 'month',
    'Dy': 'day'
}))
df.set_index('date', inplace=True)


df.drop(columns=['Yr', 'Mo', 'Dy'], inplace=True)


print(df.index.dtype)
```

```python
day_stats = pd.DataFrame({
    'min': df.min(axis=1),
    'max': df.max(axis=1),
    'mean': df.mean(axis=1),
    'std': df.std(axis=1)
})
day_stats
```

```python
df = df.copy()
df = df.loc[df.index >= '1961-01-02'] 
df['week'] = df.index.to_period('W-MON').start_time

weekly_stats = df.groupby('week').apply(lambda week_df: pd.Series({
    'min': week_df.min(axis=1).mean(),
    'max': week_df.max(axis=1).mean(),
    'mean': week_df.mean(axis=1).mean(),
    'std': week_df.std(axis=1).mean()
}))
print("the number of weak:", weekly_stats.shape[0])
weekly_stats_52 = weekly_stats.head(52)
display(weekly_stats_52.head())
```
